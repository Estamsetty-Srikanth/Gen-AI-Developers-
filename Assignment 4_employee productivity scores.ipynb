```python
''''A company wants to predict employee productivity scores to improve workforce planning and training programs. 
You are hired as a Data Scientist to build a multivariate linear regression model that predicts an employeeâ€™s Productivity Score
based on multiple work-related factors. Experience (yrs),Training Hours,Working Hours,Projects,Productivity Score 
2,40,38,3,62 
5,60,42,6,78 
1,20,35,2,55 
8,80,45,8,88 
4,50,40,5,72 
10,90,48,9,92 
3,30,37,4,65 
6,70,44,7,82 
7,75,46,7,85 
2,25,36,3,60
'''
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Creating dataset
data = {
    "Experience": [2,5,1,8,4,10,3,6,7,2],
    "TrainingHours": [40,60,20,80,50,90,30,70,75,25],
    "WorkingHours": [38,42,35,45,40,48,37,44,46,36],
    "Projects": [3,6,2,8,5,9,4,7,7,3],
    "Productivity": [62,78,55,88,72,92,65,82,85,60]
}

df = pd.DataFrame(data)

# Splitting features and target
X = df[["Experience", "TrainingHours", "WorkingHours", "Projects"]]
y = df["Productivity"]

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Predict example
prediction = model.predict([[5, 65, 42, 6]])
print("Predicted Productivity:", prediction[0])
```

```python
# Which factor most strongly impacts productivity?
The factor with the largest regression coefficient has the strongest impact.
Typically, Training Hours or Experience shows the highest influence.
```

```python
# How does training affect productivity?
Training has a positive relationship with productivity.
More training hours generally increase employee performance.
```

```python
# Should the company increase training hours or working hours?
Increasing training hours is more sustainable for productivity growth.
Excessive working hours may lead to fatigue and reduced efficiency.
```

```python
# What happens if Working Hours increase beyond optimal limits?
Productivity may initially increase but can drop due to burnout.
Overworking reduces focus, motivation, and long-term performance.
```

```python
# Can productivity ever decrease with more experience?
Yes, if experience leads to complacency or lack of skill upgrades.
Without continuous learning, growth may slow down.
```

```python
# How would you detect overfitting in this model?
Compare training accuracy with test accuracy.
If training score is high but test score is low, the model is overfitting.
```
